# Briefly unitary operators

##### We have different and equivalent definitions of unitary matrices. Let $ U \in \mathbb{C}^{n \times n} $ be a unitary matrix. Then:

- $ U $ is length preserving.
- The columns of $ U $ form an orthonormal set.
- $ U^{\dagger}U=I $ and $ U^{-1} = U^{\dagger} $. (Remember that $ U^\dagger $ is the conjugate transpose of $U$.)
- The rows of $ U $ form an orthonormal set.

Remark that $ U^\dagger $ is unitary, as well, and $ U $ is a reversible operator.

## One qubit unitary operators

We review a few one-qubit unitary operators for warming up.

### Hadamard

We start with Hadamard operator.

$
    H = \hadamard.
$

It is easy to see that $H^{\dagger} = H$, and $ HH= I $. Thus, $ H $ is a unitary operator.

### Pauli matrices

Pauli matrices are well-known and fundamental single-qubit unitary operators.

There are three Pauli matrices: $ X = \X$, $ Y = \Y$, and $ Z = \Z$, which are rotations by $ \pi $ respectively over the $ x $-axis, the $ y $-axis, and the $z$-axis.

Similar to $ H $, $ X^\dagger = X $, $ Y^\dagger = Y $, and $ Z^\dagger = Z $.

## Task 1

Verify that $Y^{\dagger} = Y$ and that $YY = I$.



### Solution

$Y^{\dagger} = \Y = Y.$

$\Y \Y = \mymatrix{cc}{-i^2 & 0 \\ 0 & -i^2} = \mymatrix{cc}{1 & 0 \\ 0 & 1}$.

## Task 2

Verify that $M = \mymatrix{cc}{ \frac{1+i}{2} & \frac{1-i}{2} \\ \frac{1-i}{2} & \frac{1+i}{2} }$ is a unitary matrix.

### Solution

Let $\ket{a}$ be a qubit, so $a_0^2+a_1^2=1$.

$
Ma= \mymatrix{cc}{ \frac{1+i}{2} & \frac{1-i}{2} \\ \frac{1-i}{2} & \frac{1+i}{2} }\mymatrix{cc}{ a_0 \\ a_1 }= \mymatrix{cc}{ \frac{1+i}{2} a_0 + \frac{1-i}{2} a_1 \\ \frac{1-i}{2} a_0 + \frac{1+i}{2} a_1}
$ with length

$
\frac{1}{4}((a_0 + a_1)^2 + (a_0 - a_1)^2 + (a_0 + a_1)^2 + (-a_0 + a_1)^2)=\frac{1}{4}((a_0^2 + a_1^2+2a_0a_1) +(a_0^2 + a_1^2-2a_0a_1)+(a_0^2 + a_1^2+2a_0a_1)+(a_0^2 + a_1^2-2a_0a_1))=\frac{1}{4}4=1
$
